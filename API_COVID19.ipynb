```python

Data_info =[]
import pandas as pd
for t in dict_data_covid19:
    Data_info.append( [t['CountryCode'],t['Country'],t['CountryCode'],t['Slug'],t['NewConfirmed'],t['TotalConfirmed'],t['NewRecovered'],t['Date']])
    
```

```python
Data_info
```

```python
import pandas as pd
df1 = pd.DataFrame (data= Data_info, columns = ['CountryCode','Country','CountryCode','Slug','NewConfirmed','TotalConfirmed','NewRecovered','Date'])
```

```python
df1
```

```python

import pandas as pd
import pyodbc 
connection = pyodbc.connect('Driver={SQL Server};'
                      'Server=NANKUTA\SQLEXPRESS;'
                      'Database=COVID19;'
                      'Trusted_Connection=yes;')

cursor = connection.cursor()
```

```python
for index,row in df1.iterrows():
    cursor.execute("insert into dbo.Covid19(Country,CountryCode,Slug,TotalConfirmed,NewRecovered,Date) values (?,?,?,?,?,?)",
                   row['Country'],
                   row['CountryCode'],
                   row['Slug'],
                   row['TotalConfirmed'],
                   row['NewRecovered'],
                   row['Date']) 
    connection.commit()
cursor.close()
```
